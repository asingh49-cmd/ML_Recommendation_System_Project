```python
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings("ignore", category=DeprecationWarning)
```

```python
orders = pd.read_csv('clean_orders.csv')
orders.head()
```

```python
basket = (
    orders
    .groupby(['order_id', 'product_name'])['quantity']
    .sum()
    .unstack()
    .fillna(0)
)
```

```python
basket = basket.applymap(lambda x: 1 if x > 0 else 0).astype(bool)
```

```python
basket.head()
```

```python
basket.describe()
```

```python
from mlxtend.frequent_patterns import apriori, association_rules

min_support = 20 / basket.shape[0]
frequent_itemsets = apriori(basket,
                            min_support=min_support,
                            use_colnames=True,
                            low_memory=True)

rules = association_rules(frequent_itemsets,
                          metric="confidence",
                          min_threshold=0.01)
```

```python
rules[['antecedents','consequents','support','confidence','lift']]
```

```python
rules.to_csv('association_rules.csv', index=False)
```

```python
basket.shape
```

```python
# 1) Clean: turn frozensets into single-item strings (works for 1→1 rules)
rules_clean = rules.copy()
rules_clean = rules_clean[rules_clean['antecedents'].apply(len) == 1]
rules_clean = rules_clean[rules_clean['consequents'].apply(len) == 1]

rules_clean['product'] = rules_clean['antecedents'].apply(lambda s: next(iter(s)))
rules_clean['complement'] = rules_clean['consequents'].apply(lambda s: next(iter(s)))

# 2) Optional: keep only positive association
rules_clean = rules_clean[rules_clean['lift'] > 1].copy()

# 3) Create rank (THIS is the missing step in your notebook state)
rules_clean = rules_clean.sort_values(['product', 'lift', 'confidence', 'support'],
                                      ascending=[True, False, False, False])

rules_clean['rank'] = rules_clean.groupby('product').cumcount() + 1

# 4) Top 5
top3 = rules_clean[rules_clean['rank'] <= 3].copy()

top3_table = top3[['product', 'complement', 'lift', 'confidence', 'support', 'rank']] \
    .sort_values(['product', 'rank'])

top3_table.head(20)
```

```python
top3_table.to_csv("product_top3_complements.csv", index=False)
```

We still face the problem of cold start with new products or ones that haven't been bought enough yet. Will now try to use the product info to identify if types of products are more commonly complements, instead of checking for each product.

```python
products = pd.read_csv('clean_products.csv')
products.head()
```

```python
products_unique = products[['product_name','category_sub', 'brand']].drop_duplicates()
orders2 = orders.merge(
    products_unique,
    left_on='product_name',
    right_on='product_name',
    how='left'
)
orders2.head()
```

```python
orders.shape
```

```python
orders2.shape
```

```python
products.groupby('product_name')['category_sub'].nunique().sort_values(ascending=False).head(10)
```

```python
conflicts = (
    products
    .groupby('product_name')
    .filter(lambda x: x['category_sub'].nunique() > 1)
)
```

```python
conflict_summary = (
    conflicts
    .groupby('product_name')
    .agg({
        'brand': lambda x: x.dropna().unique().tolist(),
        'category_sub': lambda x: x.dropna().unique().tolist()
    })
    .reset_index()
)
```

```python
conflict_summary
```

```python
basket_category = (
    orders2
    .groupby(['order_id', 'category_sub'])['quantity']
    .sum()
    .unstack()
    .fillna(0)
)
basket_category = basket_category.map(lambda x: 1 if x > 0 else 0).astype(bool)

basket_brand = (
    orders2
    .groupby(['order_id', 'brand'])['quantity']
    .sum()
    .unstack()
    .fillna(0)
)
basket_brand = basket_brand.map(lambda x: 1 if x > 0 else 0).astype(bool)
```

```python
basket_category.head()
```

```python
basket_brand.head()
```

```python
min_support = 20 / basket.shape[0]
frequent_itemsets2 = apriori(basket_category,
                            min_support=min_support,
                            use_colnames=True,
                            low_memory=True)

rules_category = association_rules(frequent_itemsets2,
                          metric="confidence",
                          min_threshold=0.01)

min_support = 20 / basket.shape[0]
frequent_itemsets3 = apriori(basket_brand,
                            min_support=min_support,
                            use_colnames=True,
                            low_memory=True)

rules_brand = association_rules(frequent_itemsets3,
                          metric="confidence",
                          min_threshold=0.01)
```

```python
rules_category[['antecedents','consequents','support','confidence','lift']]
rules_brand[['antecedents','consequents','support','confidence','lift']]
rules_category.to_csv('association_rules_category.csv', index=False)
rules_brand.to_csv('association_rules_brand.csv', index=False)
```
